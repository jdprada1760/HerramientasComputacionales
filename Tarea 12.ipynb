Hace algunas clases abordaron el problema de ajuste de parámetros de un modelo teórico a un conjunto de datos.
Para esto, la idea era definir la suma de las distancia cuadráticas desde el conjunto de datos a los respectivos valores teóricos. Esta cantidad es llamada $\chi^2$ y está definida de la siguiente manera:

$$\chi^2 = \sum_{i=1}^N \frac{(f(x_i)-y_i)^2}{\sigma_i}$$

Donde $\{x_i,y_i\}$ es el conjunto de datos al que queremos ajustar una función $y = f(x)$. $\sigma_i$ es el error asociado al valor $y_i$, que en el caso experimental sería el error absoluto asociado a una medición, como han visto en Física experimental I. El error $\sigma_i$ se toma como $1$ si no es especificado.

Veamos un ejemplo. En el archivo line.txt podemos encontrar datos $\{x_i,y_i\}$ con tendencia lineal, a los cuales queremos ajustar una recta.

------------------------------

--------------------------------

````python
# Cargamos los datos
datos = np.loadtxt("line.txt")
# Obtenemos los valores de x e y
x = datos[:,0]
y = datos[:,1]
# Graficamos con scatter 
plt.scatter(x,y)
````

------------------------------

--------------------------------
Como la tendencia es lineal, la función que queremos ajustar es:

$$ f(x) = ax+b ,$$

para algún par de números $a,b$. La idea del ajuste es obtener $a,b$ que minimicen el $\chi^2$. 

Primero veamos gráficamente el $\chi^2$. Por ejemplo, queremos ver cuánto vale el $\chi^2$ para los valores específicos de $a= 6$, $b = 8$. Para eso graficamos la recta $a= 6$, $b = 8$ además de las distancias en $y$ de la recta a los datos.

------------------------------

--------------------------------

````python
# Definimos la función lineal f
def f(x,a,b):
        return a*x+b
# Graficamos los datos "experimentales"
plt.scatter(x,y)
# Graficamos los la recta de juste a=4, b=6
xx = linspace(min(x)-1,max(x)+1)
plt.plot(xx,f(xx,6,8))
# Graficamos las distancias en y de los datos a la línea
for xi,yi in zip(x,y):
    plt.plot([xi,xi],[yi,f(xi,6,8)])
````

------------------------------

--------------------------------
El $\chi^2$ es simplemente la suma de los cuadrados de las distancias graficadas. Así nos damos cuenta que entre más pequeño el $\chi^2$, mejor se ajusta la función a los datos. El cuadrado es para que estos errores no se puedan cancelar entre sí.

------------------------------

--------------------------------

Ahora, si $\chi^2$ depende de los datos $\{x_i,y_i\}$ además de los parámetros $a,b$. Pero si sólo trabajamos con un con conjunto de datos $\{x_i,y_i\}$ que no cambiará,  $\chi^2$ sólo dependerá de los parámetros $a,b$.

$$ \chi^2 = \chi^2(a,b) $$

Ahora, $\chi^2$ toma el valor de $0$ cuando la función se ajusta perfectamente a los datos y toma el valor de $\infty$ en el peor de los casos. Con esto podemos crear una función de probabilidad que nos diga qué tan probable es que los parámetros $a,b$ sean los parámetros reales asociados a los datos:

$$ P(a,b) = e^{-\chi^2(a,b)} $$

Ésta función de probabilidad es conocida como verosimilitud o likelihood. Como era de esperarse, ésta toma el valor de $0$ en el peor de los casos y el valor de $1$ cuando hemos encontrado los parámetros perfectos.

Ahora, si variamos $a,b$, obtenemos una función tridimensional dada por la tripla $\left(a,b,P(a,b)\right)$, es decir. En otras palabras, $P(a,b)$ es una función de dos variables.

Calculemos la probabilidad $P(a,b)$ para distintos valores de $(a,b)$ y grafiquemos en código de colores.

------------------------------

--------------------------------

````python
# Definimos el chi^2 pasamos como parámetros
def chi2(x,y,a,b):
    # y teórico
    yteo = f(x,a,b)
    return sum((y-yteo)**2)
# Definimos la probabilidad
def Prob(x,y,a,b):
    return np.exp(-chi2(x,y,a,b))

# Creamos dos vectores que nos variarán a y b
var_a = np.linspace(0,6,100)
var_b = np.linspace(0,20,100)
# Creamos una cuadrícula de tamaño 100x100 donde guardaremos los valores de la probabilidad P(a,b)
probabilidad = np.zeros((100,100))
# Llenamos la cuadrícula con las probabilidades correspondientes
for i in range(100):
    for j in range(100):
        probabilidad[i,j] = Prob(x,y,var_a[i],var_b[j])
# Graficamos con imshow el logaritmo de la probabilidad
plt.imshow(log(probabilidad),extent=( min(var_a),max(var_a),min(var_b),max(var_b)), aspect = "auto")
plt.colorbar()
````

------------------------------

--------------------------------
De la matriz de probabilidad podemos obtener el valor más probable para $a,b$ y definir el mejor ajuste.

------------------------------

--------------------------------


````python
# Obtenemos el valor más probable y con eso las variables más probables
ind_a,ind_b= probabilidad.argmax()%100,probabilidad.argmax()//100
a_opt = var_a[ind_a]
b_opt = var_b[ind_b]
# Graficamos el mejor ajuste
# Graficamos los datos "experimentales"
plt.scatter(x,y)
# Graficamos la mejor recta
xx = linspace(min(x)-1,max(x)+1)
plt.plot(xx,f(xx,a_opt,b_opt))
print("Los valores optimos de a,b son : ", a_opt,b_opt)
````

Muchas veces necesitamos más información además de los valores de los parámetros más probables. Por ejemplo, de una distribución de probabilidad se puede obtener la desviación estándar que es interpretada como el generador del intervalo de confianza de los parámetros obtenidos: El parámetro más probable está sujeto también a errores.

La desviación estándar es una integral sobre la distribución de probabilidad. Sin embago, si creamos puntos aleatorios que sigan la distribución de probabilidad deseada, las integrales sobre la probabilidad se vuelven sumas sobre los puntos aleatorios. Específicamente, la desviación estándar se convierte en la correspondiene desviación de los datos.
